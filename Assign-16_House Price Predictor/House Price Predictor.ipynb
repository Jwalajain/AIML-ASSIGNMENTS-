```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    "Area": [1000, 1500, 2000, 2500, 3000],
    "Bedrooms": [2, 3, 3, 4, 4],
    "Price": [3000000, 4500000, 5000000, 6500000, 7000000]
}

df = pd.DataFrame(data)

X = df[["Area", "Bedrooms"]]
y = df["Price"]

model = LinearRegression()
model.fit(X, y)

try:
    area = float(input("Enter house area: "))
    bedrooms = int(input("Enter number of bedrooms: "))

    new_house = pd.DataFrame([[area, bedrooms]], columns=["Area", "Bedrooms"])

    prediction = model.predict(new_house)

    print("Predicted Price:", int(prediction[0]))

except ValueError:
    print("⚠️ Please enter valid numeric values!")
```
