```python
//password authentication //


password = input("Create password: ")

has_upper = False
has_digit = False
has_symbol = False
for ch in password:
    if ch >= 'A' and ch <= 'Z':
        has_upper = True
    elif ch >= '0' and ch <= '9':
        has_digit = True
    elif ch == '@' or ch == '#' or ch == '$':
        has_symbol = True
if has_upper and has_digit and has_symbol:
    print("Password accepted")

    login = input("Enter password: ")
    if login == password:
        print("Access Granted")
    else:
        print("Access Denied")
else:
    print("Weak password")
```
