```python
from math import sqrt

user_A = [5, 3, 4]
user_B = [5, 2, 4]
user_C = [1, 5, 2]
def euclidean_distance(u1, u2):
    return sqrt(sum((u1[i] - u2[i])**2 for i in range(len(u1))))
dist_AB = euclidean_distance(user_A, user_B)
dist_AC = euclidean_distance(user_A, user_C)

print("Distance between A and B:", dist_AB)
print("Distance between A and C:", dist_AC)
if dist_AB < dist_AC:
    print("User B is more similar to User A")
else:
    print("User C is more similar to User A")
```
