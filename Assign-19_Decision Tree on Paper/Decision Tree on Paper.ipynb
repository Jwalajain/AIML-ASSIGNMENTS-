```python
from sklearn.tree import DecisionTreeClassifier

X = [
    [0, 0],  # Sunny, High → No
    [0, 1],  # Sunny, Normal → Yes
    [1, 0],  # Rainy, High → No
    [1, 1],  # Rainy, Normal → No
    [2, 0],  # Cloudy, High → Yes
    [2, 1]   # Cloudy, Normal → Yes
]
y = [0, 1, 0, 0, 1, 1]
model = DecisionTreeClassifier()
model.fit(X, y)
prediction = model.predict([[0, 1]])

if prediction[0] == 1:
    print("Play Outside ✅")
else:
    print("Do Not Play ❌")
```
