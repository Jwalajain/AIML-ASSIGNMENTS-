```python
import pandas as pd

df = pd.read_csv("data1.csv")

print("Top 5 Rows:")
print(df.head())

numeric_cols = df.select_dtypes(include='number')
highest_col = numeric_cols.mean().idxmax()
print("\nColumn with highest average value:", highest_col)

print("\nMissing Values in Each Column:")
print(df.isnull().sum())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\n--- Insights ---")

print("1. Dataset contains", df.shape[0], "rows and", df.shape[1], "columns.")

print("2. Column with highest average value is:", highest_col)

missing = df.isnull().sum().sum()
print("3. Total missing values in dataset:", missing)

print("4. Numeric columns:", list(numeric_cols.columns))

print("5. Dataset shows variation in values indicating potential patterns for analysis.")
```
